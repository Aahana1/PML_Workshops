```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

isigmas=[0,1,2,3]

for isigma in isigmas:
    
    f='trial_types_isigma'+str(isigma)+'.csv'
    df=pd.read_csv(f,sep=',')
    plt.figure()
    plt.plot(df['prob_left'], label = 'left')
    plt.plot(df['prob_right'], label = 'right')
    plt.ylim(0.2,0.8)
    plt.xlabel('trial index')
    plt.ylabel('probability')
    plt.legend(loc = 'upper left' )
    plt.xlim([0,100])
   


        
    plt.savefig('experiment left and right prob %d.png' % isigma)
    
```
