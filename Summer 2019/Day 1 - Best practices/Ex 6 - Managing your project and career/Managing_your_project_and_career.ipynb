# Managing your project and career
### Be quick, creative, curious, and contented by heeding the wisdom of your elders. 

#### Whether you or your advisor are ready to admit it, being a modern scientist means being a software developer. Few of our mentors are able to mentor coders effectively. After 70 years of computer science there is a lot of wisdom for you to apply, but it lies elsewhere. Much of this wisdom is surprising and appears to contradict common prejudices about the scientific method. This is because computational data analysis tactics are prolific and have many parts that are sensitively connected. It's up to you to adapt it appropriately and give it to your peers. 

#### The overwhelming consensus about the best way to be quick, creative, curious, and contented is to heed the "Agile" philosophy.
## Agile Philosophy 

#### The observation which drove the development of "agile" is that when you need to develop software it requires too much fiddling and on-the-spot problem solving to be planned for with rigid preconceptions. Try to study everything and make a plan in advance and you will eventually succeed, but at a higher cost and with more lost opportunities than you realize.

### Why focus a discussion of scientific analysis on Agile

1.	Agile emphasizes speed but improves learning, understanding, and quality, even though it’s all jumping into things without lengthy pre-analysis. It’s so rapid you may feel like an imposter, despite doing well. Therefore, it’s a great introduction to ways to manage your projects in grad school where your projects are half your life.  
    a. When a normal student gets the same result as the *fastest* student the faster student did it about 10X faster on average. ([Yale study of computer science class](https://books.google.com/books?id=93tIAwAAQBAJ&printsec=frontcover&dq=Scrum:+The+Art+of+Doing+Twice+the+Work+in+Half+the+Time&hl=en&sa=X&ved=0ahUKEwjnmYHSh5TjAhU4HDQIHdLBA6IQuwUILjAA#v=onepage&q=Yale&f=false))  
    b. When a normal **team** gets the same result as the *fastest* team the faster team does it about 2,000X faster on average. ([IBM study of diverse project types](https://books.google.com/books?id=93tIAwAAQBAJ&printsec=frontcover&dq=Scrum:+The+Art+of+Doing+Twice+the+Work+in+Half+the+Time&hl=en&sa=X&ved=0ahUKEwjnmYHSh5TjAhU4HDQIHdLBA6IQuwUILjAA#v=snippet&q=%22two%20thousand%22&f=false))
2.	It’s often counterposed with “[waterfall](https://en.wikipedia.org/wiki/Waterfall_model)” and waterfall is the name given to what I and all my scientific peers try as neophytes. Even the US-DoD adopted then abandoned waterfall. Waterfall is intuitive and agile is counter-intuitive but more effective (for western educated people anyway). This helps me discuss dos and don’ts, without nagging or making readers feel isolated.   
3.	It is the result of decades of experimentation by the software development community.  
    a.	Therefore **it provides a lot for you to read about**.  
    b.	It also has a track record of better success if implemented honestly.  



### Agile is summarized with a brief manifesto which I've translated for the PhD student researcher:
**Individuals and interactions over protocols and organization tools** Have regular discussions about your progress rather than maintaining a strict schedule or believing in time-management tools.

**Working analysis and data visualization over precise structure and methods explication** Focus on useful and interpretable plots that are easy to generate and communicate, rather than painstakingly structuring the analysis and explaining every step.

**Advisor collaboration over promises and stated limitations** Get your advisor/mentor to sit down at your computer and look at your data so that you can rapidly demonstrate things, rather than exclusively meeting with limited slides or whiteboard work and making promises or naming obstacles. You'll get more useful insights, and they'll understand more but change might be hard. 

**Responding to insight over following a plan** Data analysis projects are actually dozens of smaller algorithmic experiments nested inside of a larger scientific experiment. All that matters is that you have eliminated alternative hypotheses, leave the algorithms out of your pre-conceived protocol. 


## All modern scientists are software developers?...Really?

Not all software developers write code, some are bosses. So in that sense yes. It's not clear how many scientific papers published last year required someone to write at least one line of code, it's probably >99.9%. Most research papers involved the use of custom code for data analysis and plotting. Senior researchers who don't write any lines of code are leading projects where many lines of code are written and that code is as important to favorable outcomes as the quality of the data itself. That code is becoming increasingly complex. Waiting for someone to develop a slick software tool with a GUI (like LabView) for everything is futile. Too many tools are being developed too fast, and the data formats are too diverse. For the foreseeable future (~10 years) a research group depending on GUI software will always be behind (AI might change this).

The proliferation of code in science has created reproducibility problems. We scientists write awful code, it's kludgy, poorly documented, and is full of idiosyncrasies. Agile won't fix that unless it's your goal to fix that. **The broader point is that, for now, you are a coder and need advice intended for coders. By pulling the thread that is Agile you unravel that advice.**

# Dos and Don'ts gleaned from Agile

*	Lose the habit of planning every detail of your project in advance  
    *	Do your project in “drafts” AKA “sprints”. Do the project by redoing it over and over, learning what details matter as you go. Plan your next few days only. The only whole-project plan should be a list of flexible goals and standards for success.  
    *	Do create a data management plan.  
    *	Do keep a journal of some kind.  
*	Reject the common prejudice that you need to study project related concepts in depth before jumping in.  
    *	Practice “just in time” learning. Attempt each “sprint” with only knowledge you can gain quickly and easily. With each draft of your project you’ll have more new things to study. Your knowledge from previous drafts will have settled, and you’ll be able to comprehend increasingly difficult texts and concepts. This is also “spaced repetition”.  
    *	Do master basic coding best practices and use test-based development.  
* Don't convince yourself that methods won't work just because you tried and failed.
    * You don't know what information is actually in the data -> You don't know what's impossible.
    * You don't know how the information is represented. -> You don't know what tweaks you need.
    * Run a lot of scripts in parallel, try many things but keep focused on using insights from yours and your colleague's knowledge about the experiment where the data was collected. Grouping data, detrending data, dimensionality reduction/expansion, filtering, event detection, etc. 
    * Decide to abandon approaches based on the time you wish to spend, not because you rationalized the worthlessness of an approach. 
* Conduct your literature review phase by collecting a lot of papers about the methods you want to employ but only read them as you need them. If a paper is really interesting, feel free of course. 
* Don't work alone. If you don't have a team, create one. Maybe look outside your lab, I use forums, and friends. We should probably start a "research partners personal ads" for the WashU system. 
    * You need people with knowledge and or experience to direct you to better information. Try to avoid those who offer dogmatic constraints.
    * Don't expect hands on help unless it's an official relationship or you are willing to offer help in return.
    * Check in with someone, anyone, regularly about what you did this day, and this week. Preferably this is a colleague, but anyone who knows you can tell you when you're too rigid, or too mired in details. 
* Reject the common habit of "summative" reports to your advisor (conclusions focused).  
    * Present intermediate results rather than pressuring yourself for conclusions. As the group becomes aware of the realities of the data, goals may change. It's awful to do a lot of work that doesn't get used. 
    * If you present at a meeting every week, make it the story of your week (if your advisor resists, you can find a middle ground). Focus meetings on deciding what works/doesn't work, and what needs to be changed or accomplished next week. 
* **Don't do anything like this**:

# Agile-like cycles for the research scientist

## Pre-planning stage

* Gather questions  
    * 1+ question that’s publishable if you answer just that  
    * 2+ questions that can be combined to a publishable result  (have backup questions)
* Collect literature on those questions 
* Collect the data you have to answer them (brainstorm new sources)  
* Explore data analysis literature until you have the names of 2-3 data analysis algorithms you’re convinced may work. NO MORE THAN 3!  
    * You'll collect more later, you simply can't do a good job yet  
* Create a folder structure and naming strategy for scripts and data (model it on the code shared with a paper you like). 

## Preliminary draft
**By creating a skeleton version of your project, your project itself is the plan for your project**
* Create a fake analysis pipeline with dummy functions instead of algorithms or real functions.  
* Pick a question and focus on the end-to-end chain of scripts, data importing, and plotting.
* Create a coherent "dummy" exposition describing the project and conclusions

## Preliminary review 
* Present your plotting/expositional strategy with your advisor/group.  
The details of the chain do not need to be presented but should be discussed.  
* Prioritize a question and some algorithms to implement for the next week or two.  
* Gather literature on the algorithms and libraries you are about to employ.  


## First iteration on a component
**Jump into the project with some rough attempts just to teach yourself**
* Focus on one question and/or small set of algorithms
* Adapt examples you found online or did in the past, or otherwise implement rough attempts at the algorithms
* Focus on the interconnections between data at each stage of analysis (between your scripts or functions), get the importing right and let the algorithms be simplistic. Some algorithms may still be dummy functions.
* Ignore preprocessing and data augmentation unless that's a large problem relative to algorithm implementation.
* Use fake data or just a small set of real data.
* Study just what you need to get the examples working.

## Subsequent iterations on a component
**Refine your rough work, but not too far.**
* use small amounts of real data, not fake data.
* replace dummy functions
* Include preprocessing and data augmentation as needed. 
* run minimal hyperparameter optimization as needed.
* focus on getting visualizations that make sense and understanding the features you can see in the data
   * Publication ready figures are wasted effort
* study the algorithms more (as needed)
    * Make sure the code is well commented
    * Take notes for the eventual method's section of a paper
* Identify sources of bias, misinterpretation, or measures needed to control for coincidence
* This is where you may decide to try different algorithms or approaches

## Component reviews
* Present your visualizations relevant to the question or algorithms you focused on.
* Discuss what you learned about features extant in the data and whether questions or conclusions are possible
* Solicit ideas, alternatives, and criticisms as well as new questions and conclusions
* Decide whether to continue iterating on this component, to change focus, or to move on to wholistic iterations

## Wholistic iterations
**Tackle all the details that are still bugging you**
* You should have well integrated and well performing software for each component
* Begin running parallel search through different choices for preprocessing and algorithms.  
    * re-run hyperparameter optimizations as needed
* At some iteration use the full set of data, retaining separate cross validation and test sets as needed
* integrate visualizations into figure panels and slides in anticipation of publication
    * Publication ready figures are mandatory
* Improve speed, parallelization, professionalism and user-friendliness
* Check with non-group members about the validity of your analysis and any pitfalls

## Wholistic reviews
* Present your conclusions and focus on changes  
    * Present your alternative hypothesis comparisons and other controls  
    * Present your well-defined limitations  
* Discuss whether you should begin drafting the paper
* Consider additions, deletions, or other changes
* Do more literature review to see what's changed and what you can find now that you have new vocabulary to search with.

# Imposter Syndrome and Mental Health
## *How I learned to stop worrying and love imposter syndrome*


## Imagine being selected as the first person on Mars...

**Would you refuse to go to Mars because you've never been?**
The most common form of imposter syndrome I've encountered is being afraid to do an analysis or to try to answer a question because it requires the use of an unfamiliar tool or domain specific knowledge one lacks. Agile gives you a mindset prepared for tackling problems like this and an alternative way to express pride and perfectionism. The iterative aspect gives you a natural way to learn what you need to while making tangible progress from the start and refining ad nauseum. The collaboration and communicative aspects give a natural way to constrain your explorations and find the right information.

*Agile practices may also increase your imposter syndrome related feelings.* The central concept is to iterate through, learning as you go. So, you will often feel unqualified for the next steps. As a new scientist, you'll become qualified for each iteration at the very end and then it's time for the next. If you reject Agile and try to learn everything in advance, it will take you a long time and you will not have all the cognitive associations needed to recall information at the right time anyway. If you do practice problems and examples to build the associations, why not use your real data for practice, and why not practice on real problems? 

**Do you think you would feel like you belong on Mars?**
Science is discovery, and a discovery is just a kind of accident. By definition, if you are doing science you are doing new things. Furthermore, as a PhD student everything around you is new to you personally. It's common to feel like you don't belong. If you feel that way you are right! Few people ever make it far enough to do scientific research because it is very hard. Nobody belongs here! The complexion of science is constantly changing. Those who choose to stay are the ones who belong. Only troubled people convince themselves they know who doesn't belong. The majority of people around you are leaving it up to you to decide, and many are deciding for themselves. 

**Do you think people will ask too many Mars questions?**
You feel like you’re expected to have answers because you reported something interesting, did something new, or have passed your PhD candidacy. Screw expectations! You have questions, not answers. Enlist your partner to answer those questions within the conversation. Everyone, except troubled persons, will enjoy the collaborative spirit. 
Enjoy your ignorance, especially if you are introverted. Embrace it as a driving force to talk to people, to try things, and keep going. Humans can convert bad anxiety, being anxious about failing or missing something, to  good anxiety, being anxious to try something or discover something. 


**Now imagine what it really takes to be that first person on Mars. Did you build the rocket? Are you significantly and measurably better than the second person off the ladder? Was flying the rocket super complicated or did a computer handle most of it?**

It's also common to feel like we just got lucky, that we didn't do enough, our work wasn't interesting enough, we didn't do anything special, or we don't really understand how we got here. Whether or not this is true for oneself is irrelevant. Nobel prizes are not lifetime achievements awards, and even evolution itself does not bestow parenthood based on noble virtues. If serendipity smiled upon you and you find the light uncomfortable then take that as your que to credit those who have helped you. Distract yourself from the feeling like you don't belong by focusing on welcoming and praising others.
